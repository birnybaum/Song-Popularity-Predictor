```python
X=Song_Popularity.drop(['song_name','song_popularity','Popular'],axis=1)
y=Song_Popularity['Popular'] #Binary Class
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler() #scale to rid bias of larger feature values
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression(class_weight='balanced')
logmodel.fit(X_train,y_train)
```

```python
predictions1 = logmodel.predict(X_test)
Predictions1 = logmodel.predict_proba(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,average_precision_score
from sklearn import metrics
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(class_weight='balanced',min_samples_leaf=15)
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions2 = dtree.predict(X_test)
Predictions2=dtree.predict_proba(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600,class_weight='balanced')
```

```python
rfc.fit(X_train,y_train)
```

```python
predictions3 = rfc.predict(X_test)
Predictions3 = rfc.predict_proba(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5,weights='uniform',)
```

```python
knn.fit(X_train,y_train)
```

```python
predictions4 = knn.predict(X_test)
Predictions4 = knn.predict_proba(X_test)
```

```python
error_rate = []


for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# NOW WITH K=32
knn = KNeighborsClassifier(n_neighbors=32)

knn.fit(X_train,y_train)
predictions5= knn.predict(X_test)
Predictions5= knn.predict_proba(X_test)
```

```python
from tensorflow import keras
from keras import layers, regularizers
import scipy.io as sio
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
model = Sequential()

model.add(Dense(12,activation='relu'))

model.add(Dense(12,activation='relu'))

model.add(Dense(6,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy', metrics=["accuracy"])
```

```python
model.fit(x=X_train,y=y_train.values,
          validation_data=(X_test,y_test.values),
          callbacks=[early_stop],
          batch_size=128,epochs=200)
```

```python
trainScores=model.evaluate(X_train,y_train,verbose=0)
print("Train loss:", trainScores[0])
print("Train accuracy:", trainScores[1])
```

```python
testScores=model.evaluate(X_test,y_test,verbose=0)
print("Test loss:", testScores[0])
print("Test accuracy:", testScores[1])
P7=testScores[1]

predictions7=model.predict(X_test)


def predict_prob(number):
  return [number[0],1-number[0]]

Predictions7 = np.array(list(map(predict_prob, model.predict(X_test))))
Predictions7[:, [1, 0]] = Predictions7[:, [0, 1]]
auroc=metrics.roc_auc_score(y_test,predictions7)
```

```python
fpr={}
tpr={}
auc={}
aps={}
Predictions=[Predictions1,Predictions2,Predictions3,Predictions4,Predictions5,Predictions6,Predictions7]
names=['Logistic Classifier','Decision Tree Classifier','Random Forest Classifier',
       'K Nearest Neighbor (K=5) Classifier','K Nearest Neighbor (K=32) Classifier',
       'Support Vector Machine Classifier','Keras Classifier']
for i in range(7):
    fpr[i], tpr[i], _ = metrics.roc_curve(y_test,Predictions[i][:,1])
    auc[i] = metrics.roc_auc_score(y_test,Predictions[i][:,1])
    print('Test:',names[i])
    print()
    print('AUC = ',auc[i])
    aps[i] = average_precision_score(y_test,Predictions[i][:,1])
    print('APS = ',aps[i])
    print()
```

```python
plt.figure(figsize=(12,8))
for i in range(7):
    plt.plot(fpr[i],tpr[i],label=names[i]+' = '+str(auc[i]))
    plt.title('Area Under ROC Curve Using Predict-Proba Values')
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc=4)
```

```python
print('The largest AUC is',auc[2],'which is the',names[2])
plt.figure(figsize=(12,8))
plt.plot(fpr[2],tpr[2],label=names[2]+' = '+str(auc[2]),)
plt.title('Max AUC')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
```

```python
predictions=[predictions1,predictions2,predictions3,predictions4,predictions5,predictions6,predictions7]
for i in range(6):
    print('Test:',names[i])
    print()
    print(classification_report(y_test,predictions[i]))
    print(confusion_matrix(y_test,predictions[i]))
    print()
```
